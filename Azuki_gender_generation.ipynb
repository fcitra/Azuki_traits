```python
def check_gender(row):
  """
  This is a function to check the gender of azuki based on the hair trait.
  There are three outcomes
  F, M, and A (andro hair trait)
  
  in
    row - string, trait values of a certain token
  out
    gender - string
  """

  row = row.split(", ")

  for i in row:
    for j in female_hair_traits:
      if i == j:
        return 'F'
        break
    for k in male_hair_traits:
      if i == k:
        return 'M'
    for m in andro_hair_traits:
      if i == m:
        return 'A'
        break

  
```

```python
# Check gender of all azukis

df['Gender'] = df['Trait_list'].apply(check_gender)
```
